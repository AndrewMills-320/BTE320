In Class Assignment 2 Simple income statement calculation

```python
#Inputs:
sales = float(input("Enter sales ($): "))
cogs_pct = float(input("Enter Cost of Goods Sold (%): "))
tax_rate = float(input("Enter Tax Rate (%): "))

#Processes:
cogs = sales * cogs_pct / 100
gross_profit = sales - cogs
taxes = gross_profit * tax_rate / 100
net_income = gross_profit - taxes

#Outputs:
print("\n--- Income Statement ---")
print(f"Sales: ${sales}")
print(f"Cost of Goods Sold: ${cogs}")
print(f"Gross Profit: ${gross_profit}")
print(f"Taxes: ${taxes}")
print(f"Net Income: ${net_income}")
```

In Class Assignment 6 Rock-Paper-Scissors

```python
import random

# Ask user if they want to play
play = input("Do you want to play Rock-Paper-Scissors? (yes/no): ").lower()

while play == "yes":
    # User chooses
    user = input("Choose rock, paper, or scissors: ").lower()
    # Computer chooses
    computer = random.choice(['rock', 'paper', 'scissors'])

    print(f"Computer chose: {computer}")

    # Determine winner
    if user == computer:
        print("It's a tie!")
    elif (user == "rock" and computer == "scissors") or \
         (user == "paper" and computer == "rock") or \
         (user == "scissors" and computer == "paper"):
        print("You win!")
    else:
        print("You lose!")

    # Ask if user wants to keep playing
    play = input("Do you want to play again? (yes/no): ").lower()

print("Thanks for playing!")
```

In Class Assignment 10 Recursive function: string reversal

```python
def reverse(s):
    if len(s) > 1:
        return s[-1] + reverse(s[1:-1]) + s[0]
    else:
        return s

reverse('hello')
```

In Class Assignment 12 Loan & NPV Mini-Toolkit (Exceptions) --
Part Aâ€”Compute Simple ROI (Return on Investment)

```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
    assert isinstance(principal, (int, float)), "principal must be numeric"
    assert isinstance(annual_rate, (int, float)), "annual_rate must be numeric"
    assert isinstance(months, int), "months must be an integer"

    if principal <= 0:
        raise ValueError("Principal must be > 0.")
    if annual_rate < 0:
        raise ValueError("Annual rate cannot be negative.")
    if months <= 0:
        raise ValueError("Number of months must be > 0.")

    r = annual_rate / 12.0
    n = months
    P = float(principal)

    if r == 0.0:
        return P / n
    else:
        return r * P / (1.0 - (1.0 + r) ** (-n))


if __name__ == "__main__":
    try:
        P = float(input("Enter principal: "))
        apr = float(input("Enter annual interest rate as a decimal: "))
        n = int(input("Enter number of months: "))
        payment = loan_payment(P, apr, n)
        print(f"Monthly payment: {payment:.2f}")
    except Exception as e:
        print(e)
```

```python
def roi(cost: float, revenue: float) -> float:
    assert cost > 0, "Cost must be positive"
    assert isinstance(cost, (int, float)), "Cost must be numerical"
    assert isinstance(revenue, (int, float)), "Revenue must be numerical"

    return (revenue - cost) / cost


if __name__ == "__main__":
    c = float(input("Enter cost: "))
    r = float(input("Enter revenue: "))
    ret = roi(c, r)
    print(ret)
```

In Class Assignment 15 Classes Point & Rectangle

```python
class Point:
  def __init__(self,x,y):
    self.x = x
    self.y = y

  def distance(self, other):
    return ((self.x - other.x)**2+(self.y - other.y)**2)**0.5
```

```python
class Rectangle:
  def __init__(self):
    self.p1 = Point(0,0)
    self.p2 = Point(1,0)
    self.p3 = Point(0,1)
    self.p4 = Point(1,1)
    self.width = self.p1.distance(self.p2)
    self.height = self.p3.distance(self.p4)

  def area(self):
    return self.height * self.width

square = Rectangle()
square.area()
```
